```python
import numpy as np
import pandas as pd
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
import random as rand
```

```python
df = pd.read_csv("9with_null_model.csv")
G1 = nx.read_gexf("graph/G1.gexf")
G1_null = nx.read_gexf("graph/G1_null.gexf")
```

### in- and out-degree distribution

```python
plt.hist(df['indegree'])
plt.xlabel('indegree')
plt.ylabel('freqeuncy')
plt.title('indegree distribution from original model')
plt.show()
```

```python
plt.hist(df['null_indegree'])
plt.xlabel('indegree from null model')
plt.ylabel('freqeuncy')
plt.title('indegree distribution from null model')
plt.show()
```

```python
plt.hist(df['outdegree'])
plt.xlabel('outdegree')
plt.ylabel('freqeuncy')
plt.title('outdegree distribution from original model')
plt.show()
```

```python
plt.hist(df['null_outdegree'])
plt.xlabel('outdegree from the null model')
plt.ylabel('freqeuncy')
plt.title('outdegree distribution from null model')
plt.show()
```

### indegree vs. log(n_killed)

```python
plt.scatter(df['indegree'], df['log_n_killed'])
plt.xlabel('indegree')
plt.ylabel('log(n_killed)')
plt.show()
```

```python
plt.scatter(df['null_indegree'], df['log_n_killed'])
plt.xlabel('indegree from null model')
plt.ylabel('log(n_killed)')
plt.show()
```

```python
print(
    'correlation between log(the number of killed victims) and # of indegree in the original model \n',
    'r =', df['log_n_killed'].corr(df['indegree']), '\n',
    'correlation between log(the number of killed victims) and # of intdegree in the null model \n',
    'r =', df['log_n_killed'].corr(df['null_indegree'])
)
```
