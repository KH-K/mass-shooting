# Merge
In this notebook, I merged two different datasets: MotherJones and GunViolenceArchive.

Merging data is processed in terms of GunViolenceArchive dataset.

```python
import pandas as pd
```

```python
MJ = pd.read_csv("MJ/MotherJones.csv")
df = pd.read_csv("4coordinate.csv", index_col=0)
```

```python
df.date = pd.to_datetime(df.date)
df = df.set_index('date')

MJ.date = pd.to_datetime(MJ.date)
MJ = MJ.set_index('date')
```

```python
df['location'] = df['city'] + ", " + df['state']
df.head()
```

```python
MJ.head()
```

## Rule
From several trials, I decided to use location and date to merging data.   
Coordinates in GunViolenceArchive dataset retrieved from Googlemaps is not perfectly matched with MotherJones dataset. (in three decimal points, it works but it is better to use location -- city + state.)

```python
new = pd.merge(df, MJ, how = 'left', on = ['location', 'date'])
```

```python
new.head()
```

```python
new.columns
```

## Remove variables  
Removing some overlapping or useless variables.

```python
del new['fatalities']
del new['injured']
del new['latitude']
del new['longitude']
del new['year']
del new['sources']
del new['location.1'] 
del new['total_victims']
del new['street']
del new['city']
del new['location']
```
