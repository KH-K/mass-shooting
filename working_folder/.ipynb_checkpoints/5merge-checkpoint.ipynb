# Merge
In this notebook, I merged two different datasets: MotherJones and GunViolenceArchive.

Merging data is processed in terms of GunViolenceArchive dataset.

```
import pandas as pd
```

```
MJ = pd.read_csv("/content/drive/My Drive/Project/data_historical/MJ/MotherJones.csv")
df = pd.read_csv("/content/drive/My Drive/Project/data_historical/4coordinate.csv", index_col=0)
```

```
df.date = pd.to_datetime(df.date)
df = df.set_index('date')

MJ.date = pd.to_datetime(MJ.date)
MJ = MJ.set_index('date')
```

```
df['location'] = df['city'] + ", " + df['state']
df.head()
```

```
MJ.head()
```

## Rule
From several trials, I decided to use location and date to merging data.   
Coordinates of GunViolenceArchive dataset from Googlemaps is not perfectly matched with MotherJones dataset. (in three decimal points, it works but it is better to use location -- city + state.)

```
new = pd.merge(df, MJ, how = 'left', on = ['location', 'date'])
```

```
new.head()
```

## delete variables  
I removed some overlapping or useless variables.

```
del new['fatalities']
del new['injured']
del new['latitude']
del new['longitude']
del new['year']
del new['sources']
del new['location.1'] 
del new['total_victims']
del new['street']
del new['city']
del new['location']
```

```
new.head()
```

```
new.to_csv('new.csv')
new.to_csv('/content/drive/My Drive/Project/data_historical/5merge.csv')
MJ.to_csv('/content/drive/My Drive/Project/data_historical/MJ/MJ.csv')
```
