```python
import numpy as np 
import pandas as pd 


import os
        
import seaborn as sns
import matplotlib.pyplot as plt

import folium 
from folium import plugins
# from folium.plugins import HeatMap

import json 
from datetime import datetime

import warnings
warnings.filterwarnings("ignore")
```

```python
path = "~/19Fall/MassShooting/working_folder/"
```

```python
df = pd.read_csv(path + "7logarithm_transformation.csv", parse_dates=['date'])
print(df.columns)
```

```python
coordinate = df[['ind', 'LON', 'LAT', 'date', 'p']]
```

```python
coordinate.head(5)
```

```python
m = folium.Map(location=[40.879102, -90.459067],
              zoom_start = 4,
              tiles = 'Cartodb Positron')

for item in coordinate.index:
    lat = coordinate.loc[item, 'LAT']
    lon = coordinate.loc[item, 'LON']
    folium.CircleMarker([lat,lon],
                        radius = coordinate.loc[item, 'p'],
                        color = 'blue',
                        fill = True).add_to(m)

m.save('maps/dot_map.html')

m
```

```python
from folium.plugins import HeatMap

heatmap = folium.Map(location=[40.879102, -90.459067],
                    zoom_start = 4,
                    tiles = 'Cartodb Positron')
h = pd.DataFrame(index = range(len(df)), columns=['coordinate'])

latlon = []
for i in range(len(df)):
    lat = df['LAT'][i]
    lon = df['LON'][i]
    coor = [lat, lon]
    latlon.append(coor)
    
h['coordinate'] = latlon

HeatMap(h).add_to(heatmap)

heatmap.save('maps/heatmap.html')
heatmap
```

```python
from datetime import datetime, timedelta
```

```python
time_index = [
    (datetime.now() + k * timedelta(1)).strftime('%Y-%m-%d') for
    k in range(len(df))
]
```

```python
from folium.plugins import HeatMap

t_heatmap = folium.Map(location=[40.879102, -90.459067],
                    zoom_start = 4,
                    tiles = 'Cartodb Positron')

df['LAT'] = df['LAT'].astype(float)
df['LON'] = df['LON'].astype(float)

latlon_t = df[['LAT','LON']]


plugins.HeatMapWithTime(latlon_t,
                        index = time_index,
                        radius=20 ).add_to(t_heatmap)


t_heatmap.save('t_heatmap.html')
t_heatmap
```
