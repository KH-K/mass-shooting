```python
import json
import pandas as pd
import numpy as np
```

```python
tweets = []
for line in open('part-m-00000', 'r'):
    tweets.append(json.loads(line))
```

```python
tweets = dict(zip(range(len(tweets)), tweets))
```

```python
data1 = pd.DataFrame.from_dict(tweets)
```
